```python
df=pd.read_csv('//content//Quote-Equity-TCS-EQ-24-08-2022-to-24-08-2023.csv')
```

```python
df
```

```python
df["Date "]=pd.to_datetime(df["Date "])
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig=go.Figure(go.Scatter(x=df['Date '], y=df['OPEN ']))
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label='One Month', step="month", stepmode="backward"),
                dict(count=6, label='Six Months', step="month", stepmode="backward"),
                dict(count=1, label='One Year', step="year", stepmode="backward"),
                dict(count=1, label='Current-Year', step="year", stepmode="todate")
            ]

            )
        )
    )
)
fig.show()
```

```python
fig=go.Figure(go.Scatter(x=df['Date '], y=df['ltp ']))
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label='Current Month', step="month", stepmode="todate"),
                dict(count=21, label='Past 3 weeks', step="day", stepmode="todate"),
                dict(count=2, label='Current Quarter', step="month", stepmode="todate"),
                dict(count=3, label='Past 3 months', step="month", stepmode="backward")
            ]

            )
        )
    )
)
fig.show()
```

```python
fig=go.Figure(go.Candlestick(x=df['Date '], open=df['OPEN '], high=df['HIGH '], \
                             low=df['LOW '], close=df['PREV. CLOSE ']))
fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```
